```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("age_height.txt",sep=",")
df.head()
```

```python
# store the data in form of dependent and independent variable separately
X=df.iloc[:,0:1].values
y=df.iloc[:,1].values
```

```python
# split the dataset into training ansd test dataset
from sklearn.model_selection import train_test_split
X_trn,X_tst,y_trn,y_tst=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# fit the simple linear regression model
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_trn,y_trn)
```

```python
# visualize the linear regression result
plt.scatter(X_trn,y_trn,color='green')
plt.plot(X_trn,linreg.predict(X_trn),color='blue')
plt.title('Linear Regression')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

```python
# add polynomial term to the equation / model
from sklearn.preprocessing import PolynomialFeatures

polynom=PolynomialFeatures(degree=3)
X_polynom=polynom.fit_transform(X_trn)
```

```python
# fit the polynomial regression model
polyreg=LinearRegression()
polyreg.fit(X_polynom,y_trn)
```

```python
# visualize the polynomial regression results
plt.scatter(X_trn,y_trn,color='green')
plt.plot(X_trn,polyreg.predict(polynom.fit_transform(X_trn)),color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Age')
plt.ylabel('Height')

plt.show()
```
